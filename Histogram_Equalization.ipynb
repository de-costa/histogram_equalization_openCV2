```python
import cv2
img=cv2.imread("test5.webp")
imgray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
print(img.shape)
```

```python
print(imgray.shape)
```

```python
flatten = imgray.flatten()
```

```python
import matplotlib.pyplot as plt
plt.hist(imgray.flatten(),bins=100)
plt.xlabel('pixel value')
plt.ylabel('occurence')
plt.show()
```

```python
equ=cv2.equalizeHist(imgray)
```

```python
cv2.imshow('original image',img)
cv2.imshow("Histogram Equalization",equ)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
